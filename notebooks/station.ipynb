```python
import pandas as pd
import numpy as np
import json
import os
from typing import Dict
```

```python
instance_dir = "../data/processed"
notebook_dir = os.getcwd()
```

```python
instance_path = os.path.abspath(os.path.join(notebook_dir, instance_dir))
```

```python
def import_instance(path) -> Dict[str, np.ndarray | int]:
    try:
        with open(path, "r") as file:
            data = json.load(file)
            return data
    except FileNotFoundError:
        raise FileNotFoundError("File not found!")
    except Exception as e:
        raise e
```

```python
bari30 = import_instance(f"{instance_path}/Bari/1Bari30.json")
bari30["distance_matrix"] = np.array(bari30["distance_matrix"])
```
